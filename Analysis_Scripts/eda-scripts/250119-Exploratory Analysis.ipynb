ExportedFishTracks (targets).csv mainly saves variables that describes targets in following aspects:
1. Target strength: compensated and uncompensated strength, range
2. Geometry: minor axis of the target's signature, length, orientation, midtpoints, perimeter, range
3. Ping info: sequatial number, ID, ping date and time

Based on above information, there is a hierarchical relationship between region, process, ping, and target.
A region can run many processes, each process can blast many pings (individual accoustic signals sent out by sonar system). One ping can get one target reponse, including TS_comp, TS_uncomp, Target_compactness, Target_intensity_variations.
